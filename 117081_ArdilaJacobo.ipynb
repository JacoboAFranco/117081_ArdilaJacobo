```python
# -*- coding: utf-8 -*-
#EJERCICIO 4 TALLER
'''INICIO'''
#%% ENTRADA DE DATOS

def number():
    while True:
        try:
            a = int(input('Escriba una cantidad entera: '))
            if a>0:
                break
        except:
            print('Caracter ingresado no es correcto')
    return a 
#%% FUNCIONES PARA ANALIZAR CADA DIGITO Y SUMAR LOS QUE SON PARES
    
def searching_pair_numbers_partI(a):
    number = []
    last = 0
    while a>0:
        last  = a%10
        number.append(last)
        a = a//10
    return number

def searching_pair_numbers_partII(number):
    pairs = 0
    for i in number:
        if i%2==0:  #<----- Una forma de encontrar un numero par.
            pairs = pairs + i #<--- Sumamos los números que sean pares...
    return pairs              #... si cumple la anterior condición

#%%LECTURA DE DATOS
x = number()
y = searching_pair_numbers_partI(x)
z = searching_pair_numbers_partII(y)
print('La sumatoria de los números pares (2n) es = ', z)
if z == 0:
    print('No hay números pares')
'''FIN'''
```

```python
# -*- coding: utf-8 -*-
#EJERCICIO 5 TALLER
'''INICIO'''
#%%ENTRADA DE DATOS
def number():
    while True:
        try:
            a = int(input(' Escriba un número entero: '))
            break
        except:
            print('¡ERROR! Valor incorrecto')
    return a
#%%FUNCIONES PARA ANALIZAR EL NÚMERO Y HALLAR 5 CONSECUTIVOS

def searching_sequential_5_partI(a):
    separation = []
    last = 0
    while a > 0:
        last = a%10
        separation.append(last)
        a = a//10
    return separation

def searching_sequential_5_partII(separation):
    elements = len(separation)
    return elements

def searching_sequential_5_partIII(separation, elements):
    sequential = 0
    place = 1
    for i in separation:
        if place<elements:
            if i==5:
                if i==separation[place]:
                    sequential = sequential + 1
                    place = place + 1
            else: place = place + +1
        else: place = place + 1
    return sequential
#%% SALIDA DE DATOS (LECTURA)

x = number()
y = searching_sequential_5_partI(x)
z = searching_sequential_5_partII(y)
w = searching_sequential_5_partIII(y,z)
print(' Cinco consecutivos = ', w)
```

```python
#%%EJERCICIO 6 TALLER
'''INICIO'''
#%% ENTRADA DE DATOS
def number():
    while True:
        try:
            a = int(input('Ingresar un valor: '))
            break
        except:
            print('¡ERROR! Valor incorrecto ')
    return a

#%% FUNCIONES PARA SABER SI EL NÚMERO ES PARTE DE LA SERIE FIBONACCI

def summation(a):
    summation1 = 0
    while a >0:
        last = a%10
        summation1 += last
        a = a//10
    return summation1

def fibonacci_serie(summation1):
    fibonacci =[]
    x1 = 0 #Definimos la primera posicion que en realidad es la 0
    fibonacci.append(x1)
    x2 = 1
    fibonacci.append(x2)
    counter = 1
    for element in fibonacci:
        if counter <= summation1:
            nextt = fibonacci[counter]# <-- Se hace un indexado para la 1 pos.
            nexttt = element + nextt #<-- le sumamos la pos 1 para obt. pos. 2
            fibonacci.append(nexttt) #<-- se almacena en la lista def por fib..
            counter += 1 #se comienza a sumar desde 1 para que empiece a dar
        else:             #la serie
            break
    return fibonacci

def verification(fibonacci, summation1):
    staff = 0
    if summation1 in fibonacci:
        staff = True
    else: 
        staff = False
    return staff

#%% SALIDA DE DATOS (LECTURA)

x = number()
y = summation(x)
z = fibonacci_serie(y)
w = verification(z,y)

if w ==True:
    print(' Número pertenece a Fibonacci ')
if w == False:
    print(' Número no pertenece a Fibonacci')

'''FIN'''
```

```python
# -*- coding: utf-8 -*-
#%% EJERCICIO 7 TALLER
'''INICIO'''
#%% ENTRADA DE DATOS
def number():
    while True:
        try:
            a = int(input('Escriba una cantidad entera: '))
            if a>0:
                break
        except:
            print('Caracter ingresado no es correcto')
    return a 

#%% FUNCIONES PARA ANALIZAR EL NUMERO Y REPORTAR SI POSEE #S POSITIVOS

def repeated(a):
    numbers=[]
    last = 0
    while a >0:
        last = a%10
        numbers.append(last)
        a = a//10
    return numbers

def who(numbers):
    repeatedd = []
    only = []
    for element in numbers:
        if element not in only:
            only.append(element)
        else:
            if element not in repeatedd:
                repeatedd.append(element)
    return repeatedd

#%%% SALIDA DATOS (LECTURA)
x = number()
y = repeated(x)
z = who(y)

print('Los números que se repiten son: ', z)

'''FIN'''
```

```python
# -*- coding: utf-8 -*-
#%% EJERCICIO 8 TALLER
''' INICIO '''
#%% FUNCIONES PARA DAR RANGO Y HACER LISTAS

def first_100():
    c100 = []
    for element in range(0,101): #Hasta 101 para que me incluya el 100
         c100.append(element) #Almacenamiento
    return c100

def three_four(c100):
    mthree = [] #Creamos las listas para los multiplos de 3 y de 4
    mfour = []
    for element in c100:
        if 0<element<46: #Los 15 primeros multiplos de 3 llegan hasta el 45
            if element%3 ==0:#por tanto, deben cumplir este rango
                mthree.append(element) #se almacena si cumple rango y multiplo
        if 46<element<101: #Se toma este rango ya que ya se iteró para el 3
            if element%4==0:
                mfour.append(element) #se almacena si cumple rango y multiplos
    return mthree, mfour

#%% SALIDA DE DATOS (LECTURA)
x = first_100()
y = three_four(x)
print(y) #imprimimos

'''FIN'''
```
